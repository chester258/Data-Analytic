```python
#Extraeremos los datos de yahoo finance del SP500 y sus sectores, haremos algunas
# transformaciones, y añadieremos algunas columnas que necesitaremos para mas adelante
#hacer un analisis con power BI.
```

```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
tikers = ['^GSPC','XLE','XLB','XLI','XLY','XLP','XLV','XLF','XLK','XLC','XLU']
```

```python
for i in tikers:
    sp = yf.download(i,start="1990-01-01",end='2022-01-01')
    sp.reset_index(inplace=True)
    
    #Calculamos el retorno diaria
    sp['Daily Return'] = sp['Close'].pct_change()

    # Calcular la media móvil de 100 y 300 dias
    sp['SMA_100'] = sp['Close'].rolling(window=100).mean()
    sp['SMA_300'] = sp['Close'].rolling(window=300).mean()

    # Calculamos el retorno diario  
    sp['cumulative_return'] = (1 + sp['Daily Return']).cumprod() - 1
    
    
    # Calcular el drawdown desde el último máximo
    sp["cummax"] = sp["Close"].cummax()
    sp["drawdown"] = ((sp["cummax"] - sp["Close"]) / sp["cummax"])*100

    
    sp.to_csv(f'{i}.csv')
    

```
